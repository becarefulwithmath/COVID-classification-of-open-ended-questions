```python
import pandas as pd
import re

##/content/drive/Shareddrives/Koronawirus/studies/3 szczepionka/classification of open ended questions/ 
#/content/drive/Shareddrives/Koronawirus/studies/3 szczepionka/classification of open ended questions/open_questions_categories.xlsx
#/content/drive/Shareddrives/Koronawirus/studies/3 szczepionka/classification of open ended questions/keywords.xlsx
#3 параметра: куда записывать результат, файл для категоризирования, файл с кейвордс


#параментр какой лист из экселя брать для анализа
field_to_analyse = "główne powody" #параметр какое поле категоризировать

df = pd.read_csv('sheet.csv', usecols=[0,2,3]) # Columns by number #+параметр какой лист брать
dic = {'ą':'a', 'ć':'c', 'ę':'e', 'ł':'l', 'ń':'n', 'ó':'o', 'ś':'s', 'ź':'z', 'ż':'z'} 

df1 = pd.read_csv('keywords.csv', usecols=[0,2,4]) # File with keywords #+параметр какой лист брать
df.replace(dic, regex=True, inplace=True) #чистит весь датафрейм
df[field_to_analyse] = df[field_to_analyse].str.lower()
df[field_to_analyse] = df[field_to_analyse].str.replace(r'"', '')

df["sum"] = "" 

for index, row in df.iterrows():  # For each record x160
  for index1, row1 in df1.iterrows(): # For each category x22
    if not row1[0] in (None, ""):
      keywrds = row1['keywords'].split(";")
      for i in keywrds:
        if ('+' in i):
          ls = i.split('+')
          c = str(all(word in str(row[field_to_analyse]) for word in ls))
        else:
          c = str(i in str(row[field_to_analyse]))
        if c != "True":
          continue
        df.loc[index, 'sum'] += c
        df.loc[index, 'sum'] += ' ' #в результате разделитель должне быть ; а не новая строка
        dic1 = {'True' : row1['category_title'], 'False' : ''}
        df.replace(dic1, regex=True, inplace=True) 
        break

#результат: две колонки - respondent_id, classification result
#финальный файл называется так же как и класифицируемая колонка
df.to_csv(r'Name1.csv')
```
