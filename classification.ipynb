**Classification of field "why"** wave1 data 'data Ariadna wave2 n=3106'

```python
import pandas as pd
import re

book_path = '/content/drive/Shareddrives/Koronawirus/studies/3 szczepionka/classification of open ended questions/open_questions_categories.xlsx'
book_sheet = 'data Ariadna wave2 n=3106'

keyword_path = '/content/drive/Shareddrives/Koronawirus/studies/3 szczepionka/classification of open ended questions/keywords.xlsx'
keyword_sheet_why = 'why'

field_to_analyse_why = 'główne powody'

final_csv_full = '/content/drive/Shareddrives/Koronawirus/studies/3 szczepionka/classification of open ended questions/'+book_sheet+'_'+keyword_sheet_why+'_final_full.csv'
final_csv_short = '/content/drive/Shareddrives/Koronawirus/studies/3 szczepionka/classification of open ended questions/'+book_sheet+'_'+keyword_sheet_why+'_final_short.csv'

df =  pd.read_excel(open(book_path, 'rb'), sheet_name=book_sheet)

#data cleaning part, replacing Polish letters
dic = {'ą':'a', 'ć':'c', 'ę':'e', 'ł':'l', 'ń':'n', 'ó':'o', 'ś':'s', 'ź':'z', 'ż':'z','Ą':'a', 'Ć':'c', 'Ę':'e', 'Ł':'l', 'Ń':'n', 'Ó':'o', 'Ś':'s', 'Ź':'z', 'Ż':'z'} 
df.replace(dic, regex=True, inplace=True)
#data cleaning part, lower case, removing ''
df[field_to_analyse_why] = df[field_to_analyse_why].str.lower()
df[field_to_analyse_why] = df[field_to_analyse_why].str.replace(r'"', '')

df["sum"] = ""

df1=pd.read_excel(open(keyword_path, 'rb'), sheet_name = keyword_sheet_why)
#df2=pd.read_excel(open(keyword_path, 'rb'), sheet_name = keyword_sheet_who)

for index, row in df.iterrows(): 
  for index1, row1 in df1.iterrows(): 
    if not row1[0] in (None, ""): #the 1st column should be non-empty
      if row['y/n'] == row1['decision']: #comparing vaccination decision with decision in keywords
        keywrds = row1['keywords'].split(";")
        for i in keywrds:
          if ('+' in i):
            ls = i.split('+')
            c = str(all(word in str(row[field_to_analyse_why]) for word in ls))
          else:
            c = str(i in str(row[field_to_analyse_why]))
          if c != "True":
            continue
          df.loc[index, 'sum'] += c
          df.loc[index, 'sum'] += ' '
          dic1 = {'True' : row1['category_title'], 'False' : ''}
          df.replace(dic1, regex=True, inplace=True) 
          break

#field_to_keep = 'respondent_id' #which columns to keep
#cols_to_keep = [field_to_keep, "sum"] #which columns to keep

#final output: short
#df.loc[:, cols_to_keep].to_csv(final_csv_short)
#for debugging:
df.to_csv(final_csv_full) 
```
