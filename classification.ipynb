```python
import pandas as pd
import re
import time

timestr = time.strftime("%Y%m%d-%H%M%S")
 
book_path = '/content/drive/MyDrive/Work/Raman/8.04/test data.xlsx'
book_sheet = 'test data'
 
keyword_path = '/content/drive/MyDrive/Work/Raman/8.04/keywords.xlsx'
keyword_sheet_why = 'why'
 
field_to_analyse_why = 'why'
 
final_csv_full = '/content/drive/MyDrive/Work/Raman/8.04/' + timestr + '_' + book_sheet + '_' + keyword_sheet_why + '_final_full.csv'
final_csv_short = '/content/drive/MyDrive/Work/Raman/8.04/' + timestr + '_' + book_sheet + '_' + keyword_sheet_why + '_final_short.csv'

df =  pd.read_excel(open(book_path, 'rb'), sheet_name=book_sheet)

### START CLEAN ###
def data_clean(df, field):
  dic = {'ą':'a', 'ć':'c', 'ę':'e', 'ł':'l', 'ń':'n', 'ó':'o', 'ś':'s', 'ź':'z',
         'ż':'z','Ą':'a', 'Ć':'c', 'Ę':'e', 'Ł':'l', 'Ń':'n', 'Ó':'o', 'Ś':'s', 
         'Ź':'z', 'Ż':'z', '"':'', ',':' ', '!':' ', ';':' ', ':':' '}  
  df.replace(dic, regex = True, inplace = True)
  df.replace({'.':' ', '?':' '}, regex = False, inplace = True) #Theese symbols are regex-sensetive
  df[field] = df[field].str.lower()
  return df

def data_trim(df, field):
  df[field] = df[field].str.strip()
  df.replace("  ", " ", regex=True, inplace=True)
  return df 

df = data_clean(df, field_to_analyse_why)
df = data_trim(df, field_to_analyse_why)
### END CLEAN ###

df['sum'] = ""

### add note onn the basis on min(lenght) ###
#задается парамент min_lenght = 5. 
#если длина чистого текста меньше чем min_lenght - добавить запись в колонку “notes” (нужно ее создать) комент “very short text”
#Если to_delete_short равен 1 то удалить строки с коментом “very short text”. 

df['notes'] = ""
to_delete_short = True #parametr
min_lenght = 5 #parametr
for index, row in df.iterrows(): 
  if len(str(row[field_to_analyse_why])) <= min_lenght:
    df.loc[index, 'notes'] += 'very short text'
  if to_delete_short == True and df.loc[index, 'notes'] == 'very short text':
    df.drop(index, inplace=True)
### END ###

special_category = 'just_yes'
 
df1 = pd.read_excel(open(keyword_path, 'rb'), sheet_name = keyword_sheet_why)
 
for index, row in df.iterrows(): 

  ### Проверка существует ли колонка “notes”, если да то дописать(!) непустые записи из колонки “notes” в колонку “sum”###
  if pd.isna(df.loc[index, 'notes']) != True: 
    df.loc[index, 'sum'] += df.loc[index, 'notes']
    df.loc[index, 'sum'] += ' '
      
  ### END ###

  ### если очищенный ответ равен “” или “ ” (пробел) то присвоить категорию “just_yes” или “just_no” в зависимости от значения колонки “y/n” ###
  if str(row[field_to_analyse_why]) == '' or str(row[field_to_analyse_why]) == ' ':
    if str(row['y/n']) == 'yes':
      df.loc[index, 'sum'] += 'just_yes'
    else:
      df.loc[index, 'sum'] += 'just_no' 
  ### END ###

  for index1, row1 in df1.iterrows(): 
    if not row1[0] in (None, ""): #the 1st column should be non-empty
      if row['y/n'] == row1['decision']: #comparing vaccination decision with decision in keywords
        keywrds = row1['keywords'].split(";")
        for i in keywrds:
          if ('+' in i):
            ls = i.split('+')
            c = str(all(word in str(row[field_to_analyse_why]) for word in ls))
          else:
            c = str(i in str(row[field_to_analyse_why]))
          if c != "True":
            continue
          df.loc[index, 'sum'] += c
          df.loc[index, 'sum'] += ' '
          dic1 = {'True' : row1['category_title'], 'False' : ''}
          df.replace(dic1, regex=True, inplace=True) 
          break
  ### задается параметр special_category=”just_yes”
#Если классификация содержит в себе ”just_yes” и еще чтото, любую другую категорию - то удалить special_category из классификации для данной ячейки. 
#Пример: было “conspiracy just_yes safety_concern”, стало “conspiracy safety_concern”. Было “just_yes”, стало “just_yes”
 ###
  sum_to_list = row['sum'].split(" ")
  if 'just_yes' in sum_to_list and len(sum_to_list) > 1:
    df.loc[index, 'sum'] = df.loc[index, 'sum'].str.replace(special_category, '')
  ### end ###

field_to_keep = 'respondent_id' #which columns to keep
cols_to_keep = [field_to_keep, "sum"] #which columns to keep
 
#final output: short
#df.loc[:, cols_to_keep].to_csv(final_csv_short)
#for debugging:
df.to_csv(final_csv_full) 
```
